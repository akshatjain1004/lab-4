```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1= pd.read_csv('./college.csv')
# df1= df1.drop(columns=['Type of System'])
df1['Name']= df1['State or Union Territory']+ ' '+ df1['Type of System']
df1= df1.drop(columns=['State or Union Territory','Type of System']) 
df1
```

```python
df1_new= pd.melt(df1, id_vars=['Name'], var_name='Type', value_name='Numbers')
sns.set_theme()
sns.heatmap(df1_new.pivot('Name', 'Type', 'Numbers'), cmap='viridis', annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'})
plt.title('State Wise Colleges')
plt.xlabel('Type')
plt.ylabel('Name')
plt.show()
```

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 6))
sns.set_theme()
sns.heatmap(df1_new.pivot('Name', 'Type', 'Numbers'), cmap='viridis', annot=True, fmt='.1f', cbar_kws={'label': 'Numbers'}, ax=axs[0])
axs[0].set_title('Poster Print')
axs[0].set_xlabel('Type')
axs[0].set_ylabel('Name')

sns.heatmap(df1_new.pivot('Name', 'Type', 'Numbers'), cmap='turbo', annot=True, fmt='.1f', cbar_kws={'label': 'Numbers'}, ax=axs[1])
axs[1].set_title('Screen Display')
axs[1].set_xlabel('Type')
axs[1].set_ylabel('Name')

sns.heatmap(df1_new.pivot('Name', 'Type', 'Numbers'), cmap='inferno', annot=True, fmt='.1f', cbar_kws={'label': 'Number'}, ax=axs[2])
axs[2].set_title('Presentation')
axs[2].set_xlabel('Type')
axs[2].set_ylabel('Name')
```
