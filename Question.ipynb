```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv('./agriculture.csv')
df1
```

```python
y1= df1[df1['Particulars']=='Foodgrains']
y1= np.array(y1.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y2= df1[df1['Particulars']=='Foodgrains Rice']
y2= np.array(y2.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y3= df1[df1['Particulars']=='Foodgrains Unmilled Paddy in term of Rice']
y3= np.array(y3.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y4= df1[df1['Particulars']=='Foodgrains Wheat']
y4= np.array(y4.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y4
```

```python
df1= df1.drop(columns=['Frequency', 'Unit' ,'Particulars'])
x= list(df1.columns)
```

# First Visualisation

```python
plt.subplots(figsize=(15,10))
plt.plot(x,y1, label='Foodgrains')
plt.plot(x,y2, label='Foodgrains Rice')
plt.plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
plt.plot(x,y4, label='Foodgrains Wheat')
plt.legend()
```

```python
cmyk_cycle = plt.cycler('color', ['c', 'm', 'y', 'k'])
rgb_cycle = plt.cycler('color', ['r', 'g', 'b'])
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(8, 6))
# axes[0].set_prop_cycle(rgb_cycle)
axes[0].plot(x,y1, label='Foodgrains')
axes[0].plot(x,y2, label='Foodgrains Rice')
axes[0].plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
axes[0].plot(x,y4, label='Foodgrains Wheat')

# axes[0].set_dpi(75)
# axes[0].set_figwidth(8)
# axes[0].set_figure(figsize=(8,6))

axes[1].set_prop_cycle(cmyk_cycle)
axes[1].plot(x,y1, label='Foodgrains')
axes[1].plot(x,y2, label='Foodgrains Rice')
axes[1].plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
axes[1].plot(x,y4, label='Foodgrains Wheat')
# axes[1].set_dpi(150)
# axes[1].set_size_inches(10.24, 7.68)
axes[1].get_figure().set_size_inches(10.24, 7.68)

axes[2].set_prop_cycle(cmyk_cycle)
axes[2].plot(x,y1, label='Foodgrains')
axes[2].plot(x,y2, label='Foodgrains Rice')
axes[2].plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
axes[2].plot(x,y4, label='Foodgrains Wheat')
# axes[2].set_dpi(300)
# axes[2].set_size_inches(30, 22.5)
axes[2].get_figure().set_size_inches(30, 22.5)

axes[0].set_title('Screen Display')
axes[1].set_title('Presentation')
axes[2].set_title('Poster Print')

axes[0].legend()
axes[1].legend()
axes[2].legend()
```

## **Different Color Pallets**

```python
import seaborn as sns
df1= pd.read_csv('./agriculture.csv')
df1= df1.drop(columns=['Frequency','Unit'])
df1
```

```python
df1_new= pd.melt(df1, id_vars=['Particulars'], var_name='Month', value_name='Stocks')
sns.set_theme()
sns.heatmap(df1_new.pivot('Particulars', 'Month', 'Stocks'), cmap='viridis', annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'})
plt.title('Monthly Foodgrain Stock Heatmap')
plt.xlabel('Month')
plt.ylabel('Food Grain')
plt.show()
```

## **1. Bipolar**

```python
fig, axs = plt.subplots(1, 3, figsize=(20, 6))
sns.heatmap(df1_new.pivot('Particulars', 'Month', 'Stocks'), cmap='viridis', annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'}, ax=axs[0])
axs[0].set_title('Poster Print')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Stocks')

sns.heatmap(df1_new.pivot('Particulars', 'Month', 'Stocks'), cmap='turbo', annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'}, ax=axs[1])
axs[1].set_title('Screen Display')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Stocks')

sns.heatmap(df1_new.pivot('Particulars', 'Month', 'Stocks'), cmap='inferno', annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'}, ax=axs[2])
axs[2].set_title('Presentation')
axs[2].set_xlabel('Month')
axs[2].set_ylabel('Stocks')
```

## **Custom**

```python
from matplotlib.colors import ListedColormap
poster_cmap = ListedColormap(['#D9381E', '#009F6B', 'lightgray', 'darkgray'])
screen_cmap = ListedColormap(['blue', 'yellow', 'lightgray', 'darkgray'])
presentation_cmap = ListedColormap(['orange', '#4B0082', 'lightgray', 'darkgray'])

fig, axs = plt.subplots(1, 3, figsize=(20, 6))
sns.heatmap(df1_new.pivot('Particulars', 'Month', 'Stocks'), cmap=poster_cmap, annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'}, ax=axs[0])
axs[0].set_title('Poster Print')
axs[0].set_xlabel('Month')
axs[0].set_ylabel('Stocks')

sns.heatmap(df1_new.pivot('Particulars', 'Month', 'Stocks'), cmap=screen_cmap, annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'}, ax=axs[1])
axs[1].set_title('Screen Display')
axs[1].set_xlabel('Month')
axs[1].set_ylabel('Stocks')

sns.heatmap(df1_new.pivot('Particulars', 'Month', 'Stocks'), cmap=presentation_cmap, annot=True, fmt='.1f', cbar_kws={'label': 'Stocks'}, ax=axs[2])
axs[2].set_title('Presentation')
axs[2].set_xlabel('Month')
axs[2].set_ylabel('Stocks')
```
