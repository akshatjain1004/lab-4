```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1= pd.read_csv('./agriculture.csv')
df1
```

```python
y1= df1[df1['Particulars']=='Foodgrains']
y1= np.array(y1.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y2= df1[df1['Particulars']=='Foodgrains Rice']
y2= np.array(y2.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y3= df1[df1['Particulars']=='Foodgrains Unmilled Paddy in term of Rice']
y3= np.array(y3.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y4= df1[df1['Particulars']=='Foodgrains Wheat']
y4= np.array(y4.drop(columns=['Frequency', 'Unit' ,'Particulars'])).T
y4
```

```python
df1= df1.drop(columns=['Frequency', 'Unit' ,'Particulars'])
x= list(df1.columns)
```

# First Visualisation

```python
plt.subplots(figsize=(15,10))
plt.plot(x,y1, label='Foodgrains')
plt.plot(x,y2, label='Foodgrains Rice')
plt.plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
plt.plot(x,y4, label='Foodgrains Wheat')
plt.legend()
```

```python
cmyk_cycle = plt.cycler('color', ['c', 'm', 'y', 'k'])
rgb_cycle = plt.cycler('color', ['r', 'g', 'b'])
fig, axes = plt.subplots(nrows=1, ncols=3,figsize=(8, 6))
# axes[0].set_prop_cycle(rgb_cycle)
axes[0].plot(x,y1, label='Foodgrains')
axes[0].plot(x,y2, label='Foodgrains Rice')
axes[0].plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
axes[0].plot(x,y4, label='Foodgrains Wheat')

# axes[0].set_dpi(75)
# axes[0].set_figwidth(8)
# axes[0].set_figure(figsize=(8,6))

axes[1].set_prop_cycle(cmyk_cycle)
axes[1].plot(x,y1, label='Foodgrains')
axes[1].plot(x,y2, label='Foodgrains Rice')
axes[1].plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
axes[1].plot(x,y4, label='Foodgrains Wheat')
# axes[1].set_dpi(150)
# axes[1].set_size_inches(10.24, 7.68)
axes[1].get_figure().set_size_inches(10.24, 7.68)

axes[2].set_prop_cycle(cmyk_cycle)
axes[2].plot(x,y1, label='Foodgrains')
axes[2].plot(x,y2, label='Foodgrains Rice')
axes[2].plot(x,y3, label='Foodgrains Unmilled Paddy in term of Rice')
axes[2].plot(x,y4, label='Foodgrains Wheat')
# axes[2].set_dpi(300)
# axes[2].set_size_inches(30, 22.5)
axes[2].get_figure().set_size_inches(30, 22.5)

axes[0].set_title('Screen Display')
axes[1].set_title('Presentation')
axes[2].set_title('Poster Print')

axes[0].legend()
axes[1].legend()
axes[2].legend()
```
